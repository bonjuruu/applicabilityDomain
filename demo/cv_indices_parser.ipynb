# Demo for parsing CV-index files

```python
import os
from pathlib import Path

import pandas as pd
import numpy as np
```

```python
PATH_ROOT = Path(os.getcwd()).absolute().parent
```

```python
column_names = [f'cv{i}' for i in range(1, 6)]

path_test = os.path.join(PATH_ROOT, 'data', 'cv', 'Ames_plainCV_test.dat')
df_idx_test = pd.read_csv(path_test, header=None, sep='\s+', na_values=[0], names=column_names)
df_idx_test = df_idx_test.astype(pd.Int64Dtype())  # Handle NaN
```

```python
df_idx_test
```

```python
df_idx_test = df_idx_test - 1  # NaN will still be NaN
df_idx_test.head()
```

```python
path_output = os.path.join(PATH_ROOT, 'experiments', 'cv_indices', 'Ames_cv_test.csv')
df_idx_test.to_csv(path_output, index=False)
```

```python
path_train = os.path.join(PATH_ROOT, 'data', 'cv', 'Ames_plainCV_train.dat')
df_idx_train = pd.read_csv(path_train, header=None, sep='\s+', na_values=[0], names=column_names)
df_idx_train = df_idx_train.astype(pd.Int64Dtype())  # Handle NaN
```

```python
df_idx_train = df_idx_train - 1  # NaN will still be NaN
```

```python
df_idx_train
```

```python
path_output = os.path.join(PATH_ROOT, 'experiments', 'cv_indices', 'Ames_cv_train.csv')
df_idx_train.to_csv(path_output, index=False)
```
