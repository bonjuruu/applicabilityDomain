```python
import os
from  pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
PATH_ROOT = Path(os.getcwd()).absolute().parent
print(PATH_ROOT)
```

```python
DATALIST=("Ames", "BBBP", "Cancer", "CYP1A2", "hERG", "HIV", "Liver")
```

```python
def load_data(data, path, filename=lambda x: f'{x}_maccs.csv'):
    path_data = os.path.join(path, filename(data))
    df = pd.read_csv(path_data)
    n = df.shape[0]
    plt.figure()
    ax = df['y'].plot.hist()
    r0 = len(np.where(df['y'] == 0)[0]) / n
    r1 = 1. - r0
    print('{:04.1f}/{:04.1f}'.format(r0*100, r1*100))
    plt.title('{} {:.1f}/{:.1f}'.format(data, r0*100, r1*100))
    plt.show()
```

```python
path_data = os.path.join(PATH_ROOT, 'data', 'maccs')
for d in DATALIST:
    load_data(d, path_data)
```
